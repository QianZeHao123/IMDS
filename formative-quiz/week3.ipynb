$ f(x) = \begin{cases} 
1, & \text{if } x < 0 \\
1 - x, & \text{if } 0 \leq x \leq 2 \\
-1, & \text{if } x > 2
\end{cases} $

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the piecewise function
def f(x):
    return np.piecewise(x, [x < 0, (0 <= x) & (x <= 2), x > 2], [1, lambda x: 1 - x, -1])

# Generate x values
x_values = np.linspace(-3, 3, 400)

# Calculate corresponding y values
y_values = f(x_values)

# Plot the graph
plt.plot(x_values, y_values, label='f(x)')

# Add labels and title
plt.title('Graph of f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

$ \int_{0.55}^{2}f(x)dx $

```python
from sympy import symbols, integrate, Piecewise

# Define variables and the piecewise function
x = symbols('x')
f = Piecewise((1, x < 0), (1 - x, (0 <= x) & (x <= 2)), (-1, x > 2))

# Calculate the integral
integral_result = integrate(f, (x, 0.55, 2))

# Print the result
print("∫[0.55 to 2] f(x)dx =", integral_result.evalf())
integral_result
```

## Question 2
a contour plot for the function $ f(x, y) = x^3 - y^3 $ using Python and the popular plotting library Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return x**3 - y**3

# Generate x and y values
x_values = np.linspace(-2, 2, 100)
y_values = np.linspace(-2, 2, 100)

# Create a meshgrid from x and y values
X, Y = np.meshgrid(x_values, y_values)

# Calculate the function values for each point in the meshgrid
Z = f(X, Y)

# Create a contour plot
plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of $f(x, y) = x^3 - y^3$')

# Show the plot
plt.show()
```

```python
from sympy import symbols, diff

# Define the variables
x, y = symbols('x y')

# Define the function
f = x**3 - y**3

# Calculate the partial derivatives
df_dx = diff(f, x)
df_dy = diff(f, y)

# Evaluate the derivatives at the point (-1, -1)
gradient_at_point = (df_dx.subs({x: -1, y: -1}), df_dy.subs({x: -1, y: -1}))

print("Gradient at the point (-1, -1):", gradient_at_point)
```

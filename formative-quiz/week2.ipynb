$$ f(x)=20xe^{6x} $$

```python
from sympy import symbols, diff, exp

# 定义变量和函数
x = symbols('x')
f = 20 * x * exp(6 * x)

# 求导
derivative_f = diff(f, x)

# 打印结果
print("f'(x) =", derivative_f)
derivative_f
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify

# 定义变量和函数
x = symbols('x')
f = 20 * x * exp(6 * x)

# 将SymPy表达式转换为可用于数值计算的函数
f_func = lambdify(x, f, 'numpy')

# 定义x的取值范围
x_values = np.linspace(-1, 1, 100)

# 计算函数值和导数值
y_values = f_func(x_values)
derivative_values = lambdify(x, derivative_f, 'numpy')(x_values)

# 绘制函数图像和导数图像
plt.figure(figsize=(10, 6))

# 绘制函数图像
plt.plot(x_values, y_values, label='f(x)')

# 绘制导数图像
plt.plot(x_values, derivative_values, label="f'(x)")

# 添加标签和图例
plt.title('Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# 显示图像
plt.grid(True)
plt.show()
```
